# Apply (Function)

```python
def fee_update(x):
    if x=="A.I":
        return 1200
    elif x=="AIoT":
        return 1500
    elif x=="CNC":
        return 2000
    else:
        return 0
    
df1.Course.apply(fee_update)
```

```python
df["Obtained Marks"]=df.c1+df.c2+df.c3+df.c4+df.c5
df
```

```python
df[['Course','c1','c2','c3','c4','c5']]
```

```python
#A.I->sum
#AIoT->product
#cnc ->commulative sum
#else -> 0
def custom_fn(x1,*l1):
    if x1=="A.I":
        return np.sum(l1)
    elif x1=="AIoT":
        return np.product(l1)
    elif x1=="CNC":
        return sum(np.cumsum(l1))
    else:
        return 0
```

```python
df["Check Validation"]=df[['Course','c1','c2','c3','c4','c5']].apply(lambda x:custom_fn(*x),axis=1)
```
