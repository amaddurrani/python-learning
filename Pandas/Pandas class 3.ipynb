```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({"CNIC":np.random.randint(1,20,size=(20)),
                  "Account No" : np.random.randint(1001,1020,size=(20))
                                          })
df1.head(2)
```

```python
df2=pd.DataFrame({"CNIC":np.random.randint(1,20,size=(20)),
                  "Account No" : np.random.randint(1001,1020,size=(20)),
                   "Mobile Number" : np.random.randint(2001,3000,size=(20)),
                   "DOB" : pd.date_range('1990-01-01',periods=20),
                  "Customer Name":[chr(x) for x in np.random.randint(65,91,size=(20))],
                  "Gender":np.random.choice(["M","F"],size=20)
                                          })
df2.head(2)
```

```python
print(df1.dtypes)
print(df2.dtypes)
```

```python
# left_df right_df, on=column(merge based on this)
a=df1.merge(df2, on=["CNIC","Account No"], how='left')
```

```python
df1=pd.DataFrame({"NIC":[1,3,4],"A":[1,3,23]})
df2=pd.DataFrame({"NIC":[1,2,3],"B":[1,3,23]})
df3=pd.DataFrame({"NIC":[1,4,7],"C":[1,3,23]})

df=df1.merge(df2,on="NIC",how='left')
df=df.merge(df3,on="NIC",how='left')
df
```

```python
df2=pd.DataFrame({"CNIC":np.random.randint(1,20,size=(20)),
                  "Account No" : np.random.randint(1001,1020,size=(20)),
                   "Mobile Number" : np.random.randint(2001,3000,size=(20)),
                   "DOB" : pd.date_range('1990-01-01',periods=20),
                  "Customer Name":[chr(x) for x in np.random.randint(65,91,size=(20))],
                  "Gender":np.random.choice(["M","F"],size=20)
                                          })
df2.groupby(by=["Gender"]).agg(['count'])
```
